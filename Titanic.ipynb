Separating Male, Female and Child

```python
def male_female_child(passenger):
    age, sex = passenger
    if age<16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['Person'] = titanic_df[['Age', 'Sex']].apply(male_female_child, axis = 1)
```

```python
#Visualising no. of male, female and child
sns.factorplot('Pclass', data = titanic_df, hue = 'Person', kind = 'count')
```

Visualising no. of male, female and child using a facet grid

```python
fig = sns.FacetGrid(titanic_df, hue='Sex', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0, oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df, hue='Person', aspect = 4)
fig.map(sns.kdeplot, 'Age', shade = True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0, oldest))
fig.add_legend()
```

```python
#dropping NAN values from cabin column
deck =titanic_df['Cabin'].dropna()

#preview of the deck
deck.head()
```

```python
#Grabbing first letter from the deck
levels = []
for level in deck:
    levels.append(level[0])
    
cabin_df = DataFrame(levels)
cabin_df.columns = ['cabin']
sns.factorplot('cabin', data = cabin_df, palette = 'winter', kind = 'count')
```

```python
cabin_df = cabin_df[cabin_df.cabin!='T']

sns.factorplot('cabin', data = cabin_df, palette = 'summer', kind = 'count')
```
